```python
# for one file analysis only

import pandas as pd

filename1 = '/Users/rainbow/Documents/Image_data/MERFISH_Mouse_Brain_Receptor_Map/datasets_mouse_brain_map_BrainReceptorShowcase_Slice1_Replicate1_cell_by_gene_S1R1.csv'

cell_by_gene = pd.read_csv(filename1)
#print(cell_by_gene)

cell_by_gene = cell_by_gene.rename(columns = {'Unnamed: 0': 'Cell_ID'})
#cell_by_gene

# remove the index column: Cell_ID
cell_by_gene2 = cell_by_gene.loc[:, "Oxgr1":"Blank-160"]
#cell_by_gene2

cell_by_gene2_rowsum = cell_by_gene2.sum(axis=1)
#cell_by_gene2_rowsum
#cell_by_gene2_rowsum.describe()


filename2 = '/Users/rainbow/Documents/Image_data/MERFISH_Mouse_Brain_Receptor_Map/datasets_mouse_brain_map_BrainReceptorShowcase_Slice1_Replicate1_cell_metadata_S1R1.csv'

cell_metadata = pd.read_csv(filename2)
#print(cell_metadata)

cell_metadata = cell_metadata.rename(columns = {'Unnamed: 0': 'Cell_ID'})
#cell_metadata

#cell_metadata['volume'].describe()

# transcript_by_volume_and_cell

transcript_volume_cell = cell_by_gene2_rowsum/cell_metadata['volume']
print(transcript_volume_cell)

transcript_volume_cell.describe()
```

```python
# Batch analysis

path = "/Users/rainbow/Documents/Image_data/MERFISH_Mouse_Brain_Receptor_Map"

file_path = "datasets_mouse_brain_map_BrainReceptorShowcase_Slice" 

for i in [1,2,3]:
    files_1 = path + '/' + file_path + str(i) + '_Replicate1_cell_' + 'by_gene_S' + str(i) + 'R1.csv'
    print(files_1)
    #print(files_1 == filename1)
    files_2 = path + '/' + file_path + str(i) + '_Replicate1_cell_metadata_S' + str(i) + 'R1.csv'
    print(files_2)
    #print(files_2 == filename2)
    
    cell_by_gene = pd.read_csv(files_1)
    #print(cell_by_gene)

    cell_by_gene = cell_by_gene.rename(columns = {'Unnamed: 0': 'Cell_ID'})
    #cell_by_gene

    # remove the index column: Cell_ID
    cell_by_gene2 = cell_by_gene.loc[:, "Oxgr1":"Blank-160"]
    #cell_by_gene2

    cell_by_gene2_rowsum = cell_by_gene2.sum(axis=1)
    
    cell_metadata = pd.read_csv(files_2)
    #print(cell_metadata)

    cell_metadata = cell_metadata.rename(columns = {'Unnamed: 0': 'Cell_ID'})
    #cell_metadata

    #cell_metadata['volume'].describe()

    # transcript_by_volume_and_cell

    transcript_volume_cell = cell_by_gene2_rowsum/cell_metadata['volume']
    
    print(transcript_volume_cell)

    print(transcript_volume_cell.describe())
    
    
    
    
```
